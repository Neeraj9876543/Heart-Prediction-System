```python
import keras
```

```python
from keras.models import Sequential
models=Sequential()
```

```python
from keras.layers import Dense
```

```python
import pandas as pd
dt=pd.read_csv('/content/heart.csv')
dt.head()
```

```python
dt.corr()
```

```python
dt = dt.drop(['chol','fbs'],axis=1)
```

```python
dt.head()
```

```python
x=dt.iloc[:,0:11]
y=dt['target']
```

```python
x
```

```python
y
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
print(x.shape, y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
models.add(Dense(units=11,activation='relu'))
models.add(Dense(units=5,activation='relu'))
models.add(Dense(units=1,activation='sigmoid'))
```

```python
models.build(input_shape=(None,11))
```

```python
models.summary()
```

```python
models.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
models.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
y_pred = models.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score  #Evaluation
accuracy_score(y_test,y_pred)
```

```python
models.save('heart.h5')
```
